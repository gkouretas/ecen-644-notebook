```python
import sys
import torch
sys.path.append("..")

from pytorch_utils import *
from preprocess import *
from model_container import ModelContainer
```

```python
resnet50 = ModelContainer(
    nnet = torchvision.models.segmentation.deeplabv3_resnet50(
        num_classes = 3
    )
)
```

```python
resnet50.load_loader("pickled/training_128_160.pkl")
resnet50.set_optimizer(torch.optim.Adam(params = resnet50._network.parameters(), lr = 0.01))
resnet50.set_cost(torch.nn.BCEWithLogitsLoss())
```

```python
resnet50.train(35)
```

```python
resnet50.save_model("resnet50_model")
resnet50.save_losses("resnet50_losses")
```
