```python
import sys
sys.path.append("..")

from pytorch_utils import *
from preprocess import *
from model_container import ModelContainer
```

```python
unet = ModelContainer(
    nnet = get_instrument_segmentation_model_base(
        Models.UNET
    ),
    returns_dict = False
)
```

```python
dataloader_name = "training_128_160"

from model_container import _pickle_path

if os.path.exists(
    os.path.join(_pickle_path, dataloader_name + ".pkl")
):
    unet.load_loader(os.path.join(_pickle_path, dataloader_name + ".pkl"))
else:
    input_videos, output_videos = load_endovis_videos(DatasetType.TRAINING, training_single_instruments_only = True)
        
    input_preprocessed = preprocess_source_endovis_images(input_videos, (128, 160))
    output_preprocessed = preprocess_endovis_target_images(output_videos, (128, 160))
    
    unet.set_dataset(input_preprocessed, output_preprocessed, 64)    
    unet.save_loader(dataloader_name)
    
unet.set_optimizer(torch.optim.Adam(params = unet._network.parameters(), lr = 0.01))
unet.set_cost(torch.nn.BCEWithLogitsLoss())
```
